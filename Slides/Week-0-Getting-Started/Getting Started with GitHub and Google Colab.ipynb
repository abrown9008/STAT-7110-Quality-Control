# 📄 Student Guide: Getting Started with GitHub + Google Colab
## Part 1 – Setup at the Start of the Semester

**1. Fork the Class Repository**
    - Go to my class GitHub repository [link](https://github.com/abrown9008/STAT-7110-Quality-Control).
    - Click **"Fork"** (top right) to create your own copy under your GitHub account.
    - You now have made a copy of the main repo that you can freely modify.

**2. Open Google Colab**
    - Go to https://colab.research.google.com/.
    - Click **File → Open Notebook → GitHub**.
    - Paste your fork’s GitHub URL in the search bar.
    - Open the notebook for the specific lecture/assignment you're looking for.

## Part 2 – Recommended: Creating a Branch for Each Assignment
You will create **one branch per assignment**.

**1. On GitHub (your fork)**
    - Go to your fork in GitHub.
    - Click the **branch dropdown** (default is main).
    - Type a new branch name like: `assignment-1`
    - Click **Create branch: assignment-1 from main.**

**2. In Google Colab**
    - Open the notebook for the assignment from **your new branch, not main.**
    - To do this:
        - **File → Open Notebook → GitHub → Search for your repo → Select the branch you just created**.
    - Work on your assignment normally in Colab.
    - Save regularly (**File → Save a copy in GitHub**) — make sure to select your branch (`assignment-1`) in the dropdown when saving.
    - When complete, save your final work to your **Google Drive** as well for backup.

## Part 3 – Submitting Your Assignments

1. **Finish your work in Colab** and save it to your branch (`assignment-1`) **and** to your Google Drive.
2. From either your fork on GitHub or from Google Drive, download the final notebook file (`.ipynb`).
3. Upload the notebook file to the appropriate D2L assignment submission area.

## Part 4 – Workflow for Future Assignments

- For Assignment 2:
    1. Go to your fork’s `main` branch.
    2. Pull any updates from the instructor’s repo (via **Sync Fork** in GitHub).
    3. Create a new branch `assignment-2` from `main`.
    4. Repeat steps from **Part 2 & 3.**
